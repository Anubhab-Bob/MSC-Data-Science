<b>Statistical Regularity</b> 

```python
import random
import matplotlib.pyplot as plt
```

```python
# Calculate the relative frequencies
x = []
y = []
cumulative = 0
for i in range(6000):
    x.append(i+1)
    outcome = random.randint(0,1)
    cumulative = cumulative + outcome
    y.append(cumulative/x[i])
```

```python
# Plot 1
plt.plot(x, y)
plt.title("Statistical Regularity")
plt.xlabel("No. of trials")
plt.ylabel("Relative frequency")
plt.show()
```

```python
# Plot 2
plt.plot(x, y)
plt.title("Statistical Regularity")
plt.xlabel("No. of trials")
plt.ylabel("Relative frequency")
plt.show()
```
