# Linear Regression Model using sklearn

LinearRegression fits a linear model with coefficients w = (w1, â€¦, wp) to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
```

```python
# Load dataset
dataset = datasets.load_diabetes(as_frame=True)
print(dataset.data)
print(dataset.target)
```

```python
# Show relation between different attributes
fig, axs = plt.subplots(2, 3)
axs[0, 0].scatter(dataset["age"],dataset["sex"])
axs[0, 0].set_xlabel("age")
axs[0, 0].set_ylabel("sex")
axs[0, 1].scatter(dataset["age"],dataset["bmi"])
axs[0, 1].set_xlabel("age")
axs[0, 1].set_ylabel("bmi")
axs[0, 2].scatter(dataset["age"],dataset["bp"])
axs[0, 2].set_xlabel("age")
axs[0, 2].set_ylabel("bp")
axs[1, 0].scatter(dataset["sex"],dataset["bmi"])
axs[1, 0].set_xlabel("sex")
axs[1, 0].set_ylabel("bmi")
axs[1, 1].scatter(dataset["sex"],dataset["bp"])
axs[1, 1].set_xlabel("sex")
axs[1, 1].set_ylabel("bp")
axs[1, 2].scatter(dataset["bmi"],dataset["bp"])
axs[1, 2].set_xlabel("bmi")
axs[1, 2].set_ylabel("bp")
fig.tight_layout()
plt.show()
```

```python
# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True, as_frame=True)
# Use only bmi feature
diabetes_X = np.array(diabetes_X["bmi"]).reshape(-1,1)
print(diabetes_X)
print(diabetes_y)
```

```python
# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]


# Model can also be prepared by not creating separate training/testing sets
```

```python
# Create and fit model
model = linear_model.LinearRegression(n_jobs=-1)
model.fit(diabetes_X_train,diabetes_y_train)
# Make predictions using the testing set
diabetes_y_pred = model.predict(diabetes_X_test)
r_sq = model.score(diabetes_X_test,diabetes_y_test)
# Show results
print('Coefficients : %.3f'
      % model.coef_)
# The mean squared error
print('Intercept : %.2f'
      % model.intercept_)
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination : %.3f'
      % r_sq)
```

```python
# Plot results
plt.scatter(diabetes_X_test, diabetes_y_test,  color='red')
plt.plot(diabetes_X_test, diabetes_y_pred, color='black')
plt.show()
```
